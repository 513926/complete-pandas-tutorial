```python
import numpy as np

a = df.groupby(['noc', 'discipline'])['place'].min().reset_index()

a.loc[a['place'] > 3, 'place'] = np.NaN
```

```python
new = a.pivot(index='noc', columns='discipline', values='place')
```

```python
new.loc['USA', new.columns.str.contains("Hockey",case=False)].reset_index()
```
